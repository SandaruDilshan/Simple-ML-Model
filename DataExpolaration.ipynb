# 2nd Stage in Machine Learning Process (Data Exploration)
## How to visualize Data in Python

```python
import pandas as pd
cars = pd.read_csv('data.csv')
Survay =  pd.read_csv("survay.csv")
cars.head()
```

## Create a Relationship Visualization

```python
cars.plot(kind = 'scatter', x = 'Volume', y = 'Car')
```

## Create Create Distribution Visualization

```python
cars['Volume'].plot(kind ='hist')
```

## Create Comparison Visualization (Using Box Plot)

```python
cars.pivot(columns='Car', values='Volume')
```

```python
cars.pivot(columns='Car', values='Volume').plot(kind = 'box', figsize = (12,6))
```

## Create Composition Visualization (Stacked Bar-Chart)

```python
cars.groupby('Car')['Volume'].value_counts()
```

```python
cars.groupby('Car')['Volume'].value_counts().unstack()
```

```python
cars.groupby('Car')['Volume'].value_counts().unstack().plot( kind ='bar', stacked = True, figsize = (12,6)) 
```
