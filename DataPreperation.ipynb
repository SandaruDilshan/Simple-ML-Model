## How to detect Missing Values

```python
import pandas as pd
Diabetes = pd.read_csv('Diabetes.csv')
Diabetes
```

##### Check spesific column missing vales

```python
mask = Diabetes['Serum_Insulin'].isnull()
mask
```

##### All raws without missing values in 'Serum_Insulin' 

```python
Diabetes[mask]
```

### Remove missing Values

```python
Diabetes.dropna()
```

##### Remove only raws boath missing values in Skin_Fold and Serum_Insulin

```python
diabetes = Diabetes.dropna( subset = ['Skin_Fold', 'Serum_Insulin'], how ='all')
diabetes
```

##### Drop columns where missning data

```python
Diabetes.dropna( axis =1)
```

### Resolve missing values instead of Drop values

##### Fill all the missing values in Skin_Fold using mean value of Skin_Fold

```python
mean = Diabetes['Skin_Fold'].mean()
diabetes = Diabetes.fillna({'Skin_Fold': mean})
diabetes
```

#### Fill the spesific raw missning values 
##### Select rows

```python
mask = (Diabetes['Pregnant'] == 6 ) & ( Diabetes['Diastolic_BP'] == 72.0 )
Diabetes.loc[mask, :]
```

##### Assign value as 'Serum_Insulin' =200

```python
Diabetes.loc [ mask, 'Serum_Insulin'] = 200
Diabetes[mask]
```
